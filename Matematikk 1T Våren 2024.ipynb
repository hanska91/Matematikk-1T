## Oppgave 1 - Del 1 (2 poeng)



Tom har arbeidet med trekanten ovenfor og påstår at tan u * tan v = 1

a) Vis at Tom har rett.

b) Avgjør om påstanden stemmer for alle rettvinklede trekanter med to spisse vinkler u og v.

Oppgave 1 a)



```python
import math

# Funksjon for å verifisere Toms påstand
def verifiser_toms_påstand(a, b):
    tan_u = a / b
    tan_v = b / a
    return tan_u * tan_v == 1

# Test funksjonen med forskjellige rettvinklede trekanter
test_tilfeller = [
    (3, 4),
    (5, 12),
    (7, 24),
    (8, 15)
]

resultater = [verifiser_toms_påstand(a, b) for a, b in test_tilfeller]

print(f"1 b) Toms påstand stemmer for alle testtilfeller: {all(resultater)}")
```

# Oppgave 5 - Del 1 (4 poeng)

Figuren viser grafen til en funksjon f.



a) Bestem f(x)

b) Løs ulikheten f(x) > 12

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve, Poly

# Define the points
points = [(-3, 0), (4, 0), (0, 24)]

# Fit a polynomial of degree 2 (quadratic) to the points
coefficients = np.polyfit([p[0] for p in points], [p[1] for p in points], 2)
polynomial = np.poly1d(coefficients)

# Define the function f(x)
def f(x):
    return polynomial(x)

# Find f(x) and check if f(x) > 12
x_values = np.linspace(-3, 4, 100)
f_values = f(x_values)

# Plot the polynomial and the points
plt.plot(x_values, f_values, label='f(x)')
plt.scatter([p[0] for p in points], [p[1] for p in points], color='red', label='Points')
plt.axhline(y=12, color='green', linestyle='--', label='y=12')
plt.legend()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('f(x) and f(x) > 12')

# Show x-axis and y-axis
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Limit the number of significant digits to 2
np.set_printoptions(precision=2)

plt.show()

# Solve the inequality f(x) > 12
x = symbols('x')
inequality = Eq(polynomial(x), 12)
solutions = solve(inequality, x)

# Display the polynomial function using sympy for better formatting
polynomial_sympy = Poly(polynomial, x).as_expr()
print(f"Polynomfunksjonen som passer til disse punktene: (-3, 0), (4, 0), and (0, 24) er: {polynomial_sympy}")

# Print the solutions with 2 significant figures
solutions_rounded = [round(sol.evalf(), 2) for sol in solutions]
print(f"Løsningen for ulikheten f(x) > 12, er: {solutions_rounded}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

from scipy.optimize import curve_fit
from sympy import solveset, Symbol

plt.rcParams['axes.grid'] = True
plt.rcParams['figure.figsize'] = 9, 4


Bagetter = np.array([  100,  130,  160,  175,  190,  220,  235])
Overskudd = np.array([1450, 2300, 3050, 3365, 3720, 4140, 4175])

plt.scatter(Bagetter, Overskudd, color="orange")
plt.axhline(0, color="black")
plt.xlabel("bagetter")
plt.ylabel("overskudd (kr)")
plt.show()
```

```python
def poly2(x, a, b, c):
  return a*x**2 + b*x + c

res = curve_fit(poly2, Bagetter, Overskudd)

av, bv, cv = res[0]
 
def fun(x):
  return poly2(x, av.round(2), bv.round(2), cv.round(2))

x = Symbol("x")
O = fun(x)
print(f"1a) O(x) = {O}")
```

```python
X = np.linspace(Bagetter.min()*0.8, Bagetter.max()*1.2)
plt.plot(X, fun(X), label="modell")
plt.scatter(Bagetter, Overskudd, color="orange", label="observert")
plt.axhline(0, color="black")
plt.legend()
plt.xlabel("bagetter")
plt.ylabel("overskudd (kr)")
plt.show()
print("1 a) Funksjonen ser ut til å passe bra.")
```

```python
bagetter = 0
overskudd = fun(bagetter)
 
while fun(bagetter + 1) > overskudd:
  bagetter = bagetter + 1
  overskudd = fun(bagetter)
 
print("1b) Optimal produksjon: ", bagetter)
print("1b) Optimalt overskudd: ", overskudd)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import sympy as sp

# Given data
Bagetter = np.array([100, 130, 160, 175, 190, 220, 235])
Overskudd = np.array([1450, 2300, 3050, 3365, 3720, 4140, 4175])

# Define the model function to be fitted to the data
def model(x, a, b, c):
    return a * x**2 + b * x + c

# Fit the model to the data
params, _ = curve_fit(model, Bagetter, Overskudd)

# Generate x values for plotting the fitted curve
x_fit = np.linspace(min(Bagetter), max(Bagetter), 100)
y_fit = model(x_fit, *params)

# Plot the data points and the fitted curve
plt.scatter(Bagetter, Overskudd, color="orange", label="Data points")
plt.plot(x_fit, y_fit, color="blue", label="Fitted curve")
plt.axhline(0, color="black")
plt.xlabel("Bagetter")
plt.ylabel("Overskudd (kr)")
plt.legend()
plt.show()

# Print the fitted parameters
print(f"Fitted parameters: a = {params[0]}, b = {params[1]}, c = {params[2]}")

# Calculate the instantaneous rate of change at x = 175
x = sp.Symbol('x')
f = params[0] * x**2 + params[1] * x + params[2]
f_prime = sp.diff(f, x)
instantaneous_rate_of_change = f_prime.subs(x, 235)

print(f"Bestem den momentane vekstfarten når x = 235 er {instantaneous_rate_of_change} kr")
```
